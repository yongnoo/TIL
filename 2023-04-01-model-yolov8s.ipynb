```python
test_list = glob.glob('./data/images/test/*.jpg')
```

```python
test_list_img = []
```

```python
for i in test_list :
    img = cv2.imread(i,cv2.IMREAD_COLOR)
    test_list_img.append(img)
```

```python
sample_test = random.sample(test_list_img,100)
```

```python
presult = model(sample_test)
```

```python
def write_res (presult,name) :
    for i,pre in enumerate(presult):
        print(i)
        file_name = pre.path
        box = pre.boxes
        pre_plot = pre.plot()
        to_path = 'g:/prj/yolov8/runs/detect/'+name+'/predict/'

        if not os.path.exists(to_path) :
            os.makedirs(to_path)
        cv2.imwrite(to_path+file_name,pre_plot)
```

```python
presult = model(sample_test)
write_res(presult,'custom_v8s7')
```
