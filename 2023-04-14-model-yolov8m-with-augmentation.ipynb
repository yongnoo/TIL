```python
from ultralytics import YOLO
from ultralytics.yolo.utils.torch_utils import strip_optimizer
import os
```

```python
datasets = ['train','val','test']
filetype = ['labels','images']

data_path = './data/'

for ft in filetype :
    for ds in datasets :
        path = os.path.join(data_path,ft,ds)
        print(ft,ds,len(os.listdir(path)))
```

```python
for ft in filetype :
    for ds in datasets :
        path = os.path.join(data_path,ft,'last',ds)
        print(ft,ds,len(os.listdir(path)))
```
